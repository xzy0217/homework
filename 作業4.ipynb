```python
import os
os.chdir('/content/drive/MyDrive/exam')  # Colab 換路徑使用

import openpyxl
wb = openpyxl.load_workbook('test.xlsx', data_only=True)  # 設定 data_only=True 只讀取計算後的數值

s1 = wb['csv']

def get_values(sheet):
    arr = []                      # 第一層串列
    for row in sheet:
        arr2 = []                 # 第二層串列
        for column in row:
            arr2.append(column.value)  # 寫入內容
        arr.append(arr2)
    return arr

print(get_values(s1))       # 印出工作表 1 所有內容

```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('population.csv')
data.to_excel('test.xlsx', index=False)
excel_data = pd.read_excel('test.xlsx')
plt.plot(excel_data['pop'],excel_data['year'])
plt.xlabel('pop')
plt.ylabel('year')
plt.show()
```
